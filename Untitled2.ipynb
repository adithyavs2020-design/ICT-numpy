```python
import numpy as np
import pandas as pd


df = pd.read_csv("auto-mpg.csv")


mpg = df['mpg'].values

mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_mpg = np.std(mpg)

count_above_25 = np.sum(mpg > 25)

print(mean_mpg, median_mpg, std_mpg, count_above_25)
```

```python
# 2. Filtering
filtered = df[df['cylinders'] > 6]['car name'].tolist()
print(filtered)
```

```python
# 3. Broadcasting
hp = df['horsepower'].replace('?', np.nan).astype(float)

hp_mean = np.nanmean(hp)
hp = np.where(np.isnan(hp), hp_mean, hp)

hp_updated = hp * 1.10
print(hp_updated)
```

```python
# 4. Boolean Indexing
displacement = df['displacement'].values
origin = df['origin'].values

avg_disp = np.mean(displacement[origin == 2])
print(avg_disp)
```

```python
# 5. Conditional Aggregates
cylinders = df['cylinders'].values

for cyl in np.unique(cylinders):
    print(cyl, np.mean(mpg[cylinders == cyl]))
```

```python
# 1. Basic Exploration
print(df.head(10))
print(df.shape)
print(df.describe())
```

```python
# 2. Filtering and Indexing
filtered_df = df[(df['model year'] == 75) & (df['weight'] < 3000)]
print(filtered_df[['car name', 'weight', 'mpg']])
```

```python
# 3. Handling Missing Data
df['horsepower'] = df['horsepower'].replace('?', np.nan).astype(float)
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)
```

```python
# 4. Data Transformation
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
```

```python
# 5. Group By
print(df.groupby('origin')['mpg'].mean())
```

```python
# 6. Visualization Preparation
summary = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
print(summary)
```

```python

# 7. Exporting Data
high_mpg = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg.to_csv("high_mpg_cars.csv", index=False)
```
